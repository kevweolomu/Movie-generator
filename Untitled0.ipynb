### ***Model selection and training***

```python
a = np.array([score_series]).reshape(-1,1)
b = np.array([GDP_per_capital_series]).reshape(-1,1)
c = np.array([social_support_series]).reshape(-1,1)
d = np.array([healthy_life_expectancy_series]).reshape(-1,1)
e = np.array([freedom_to_make_life_choices_series]).reshape(-1,1)
f = np.array([generosity_series]).reshape(-1,1)
g = np.array([perceptions_of_corruption_series]).reshape(-1,1)
```

```python
# spliting the data into train and test for supervised learning, specifying the split size of 70% and 30% respectively. 
# Since train_test_splits arrays or matrices into random train and test, specifying a random state controls the shuffling applied to the data before applying the split

a_train, a_test, b_train, b_test, c_train, c_test = train_test_split(a, b, c, train_size=0.7, test_size=0.3, random_state = 1)

# displaying the length/shape of the split data
print("Number of records in a_train sample:", len(a_train))
print("Number of records in a_test sample:", len(a_test))
print("")

print("Number of records in b_train sample:", len(b_train))
print("Number of records in b_test sample:", len(b_test))
print("")

print("Number of records in c_train sample:", len(c_train))
print("Number of records in c_test sample:", len(c_test))
```

```python
# the shape of the train and test data

print(a_train.shape)
print(a_test.shape)
print("")

print(b_train.shape)
print(b_test.shape)
print("")

print(c_train.shape)
print(c_test.shape)
```

### ***Linear Regression Model***

```python
model_one = LinearRegression().fit(a_train, b_train)
model_two = LinearRegression().fit(a_train, c_train)
```

```python
# score the model to check how accurate it is
r_sq_1 = model_one.score(a_test, b_test)         # Returns the coefficient of determination of the prediction
print('coefficient of determination () of model_one test set:', r_sq_1)

r_sq_11 = model_one.score(a_train, b_train)        
print('coefficient of determination () of model_one train set:', r_sq_11)
print("MAE score on model_one test set:", mean_absolute_error(a_test, b_test))
print("MAE score on model_one train set:", mean_absolute_error(a_train, b_train))
print("")

r_sq_2 = model_two.score(a_test, c_test)         
print('coefficient of determination () of model_two test set:', r_sq_2)

r_sq_22 = model_two.score(a_train, c_train)        
print('coefficient of determination () of model_two train set:', r_sq_22)
print("MAE score on model_two test set:", mean_absolute_error(a_test, c_test))
print("MAE score on model_two train set:", mean_absolute_error(a_train, c_train))
print("")

print("")
print('slope () of model_one:', model_one.coef_)  # returns a two dimensional array with a single element 
print('slope () of model_two:', model_two.coef_) 

print("")
print('intercept () of model_one:', model_one.intercept_)       #returns a one-dimensional array with a single element 
print('intercept () of model_two:', model_two.intercept_)
```

```python
# using the model to predict on the Test data 
b_test_prediction_one = model_one.predict(a_test)
print('Predicted response for model one:', b_test_prediction_one, sep='\n')
print("")

c_test_prediction_two = model_one.predict(a_test)
print('Predicted response for model two:', c_test_prediction_two, sep='\n')
```

```python
b_test_pred_one = model_one.intercept_ + np.sum(model_one.coef_ * a_test, axis=1)
print('Predicted response for model_one:', b_test_pred_one, sep='\n')
print("")

c_test_pred_two = model_two.intercept_ + np.sum(model_two.coef_ * a_test, axis=1)
print('Predicted response for model_two:', c_test_pred_two, sep='\n')
```

```python
# Validating the models for b_test_prediction_one, c_test_prediction_two

print("R^2 score on test set model_one:", r2_score(b_test, b_test_prediction_one ))
print("MAE score on test set model_one:", mean_absolute_error(b_test, b_test_prediction_one ))
print("")

print("R^2 score on test set model_two:", r2_score(c_test, c_test_prediction_two ))
print("MAE score on test set model_two:", mean_absolute_error(c_test, c_test_prediction_two ))
```

```python
scaler = StandardScaler()
```

```python
# fit the scaler with the train data and transform the train data
train = scaler.fit_transform(train)

# transform the test data
test = scaler.transform(test)
```

Since we do not know the number of the clusters (k) from the data, we can use the Elbow method to find the optimal number of clusters in the dataset.

The opitmal number of clusters(k) is plotted against the distortion which is the total of within-cluster sum of squares for a given number of k). Here, the optimal number of cluster (k) is 3. 

```python
visualizer = KElbowVisualizer(model, k=(1,16)).fit(train)
visualizer.show()
```

```python
model = KMeans(random_state=0, init="random", max_iter=2000, n_clusters=3)
model.fit(train)
```

```python
model_df = pd.DataFrame(data=train, columns=['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption'])
model_df["labels"] = model.labels_
model_df
```

```python
print(model.labels_)             # returns predicted class labels(cluster) for each data point
print("")

print("Sum of the squared distances from closest cluster:", model. inertia_)           # returns the within-cluster sum of squares. This is a total of the within-cluster sum of squares for all clusters   
print("")

print(model.n_iter_)             # returns the number of iterations that k-means algorithm ran to get a minimum within-cluster sum of squares
print("")

print(model.cluster_centers_)    #  # returns the location of the centroids on each cluster. The data point in a cluster will be close to the centroid of that cluster.
```

The k-means clustering has 3 clusters of size 56, 34, 19
